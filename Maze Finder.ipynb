```python
import os
os.chdir("C:\\Users\\b.saunders\\Documents\\Python\\Projects\\Maze")

def coord(a):
    return a[0], a[1]

def find_north(co):
    #Finds the status of the path to the north
    x_co, y_co = coord(co)[0] - 1 , coord(co)[1]
    if x_co < 0:
        x_co = 'n'
    if y_co < 0:
        y_co = 'n'    
    return x_co,y_co

def find_east(co):
    #Finds the status of the path to the East
    x_co, y_co = coord(co)[0] , coord(co)[1] + 1
    if x_co < 0:
        x_co = 'n'
    if y_co < 0:
        y_co = 'n'    
    return x_co,y_co
    
def find_south(co):
    #Finds the status of the path to the South
    x_co, y_co = coord(co)[0] + 1 , coord(co)[1] 
    if x_co < 0:
        x_co = 'n'
    if y_co < 0:
        y_co = 'n'    
    return x_co,y_co
  
def find_west(co):
    #Finds the status of the path to the West
    x_co, y_co = coord(co)[0], coord(co)[1] - 1   
    if x_co < 0:
        x_co = 'n'
    if y_co < 0:
        y_co = 'n'    
    return x_co,y_co

def return_value(maze,a):
    return maze[coord(a)[0]][coord(a)[1]]

def find_start(maze):
    #Identifies the start of the maze (Block in first row with a "0")
    first_line = maze[0]
    for i in range(len(first_line)):
        if first_line[i] == 0:
            pos = i
            break
    start = 0,pos
    return start
    
def find_end(maze):
    #Identifies the start of the maze (Block in last row with a "0")
    last_line = maze[len(maze) - 1]
    for i in range(len(last_line)):
        if last_line[i] == 0:
            pos = i
            break
    return len(maze) - 1,pos
        
def check_around(maze,a):
    #Looks around the current postion and returns values of the blocks directly around in the maze
    north = find_north(a)
    east = find_east(a)
    south = find_south(a)
    west = find_west(a)
    if north[0] == 'n' or north[1] == 'n':
        n_v = 'n'
    if north[0] != 'n' and north[1] != 'n':
        n_v = return_value(maze,north)
    if east[0] == 'n' or east[1] == 'n':
        e_v = 'n'
    if east[0] != 'n' and east[1] != 'n':
        e_v = return_value(maze,east)
    if south[0] == 'n' or south[1] == 'n':
        s_v = 'n'
    if south[0] != 'n' and south[1] != 'n':
        s_v = return_value(maze,south)
    if west[0] == 'n' or west[1] == 'n':
        w_v = 'n'
    if west[0] != 'n' and west[1] != 'n':
        w_v = return_value(maze,west)
    return n_v,e_v,s_v,w_v

def pretty(a):
    a = str(a)
    a = a.replace(" ","")
    return a

def distance(maze,a):
    #finds x and y distance to the end of the maze
    end = find_end(maze)
    x_distance = abs(a[0] - end[0])
    y_distance = abs(a[1] - end[1])
    return x_distance, y_distance

def create_dict(maze):
    #Creates a Nodes dictionary for all explored nodes in the maze, being adding to later  
    start = find_start(maze)
    nodes = {pretty(start):{'co':start,'s':1,'explored':False,'via':pretty(start),'y-distance': distance(maze,start)[0], 'x-distance': distance(maze,start)[1]}}
    return nodes

def next_move(maze,nodes,a):
    #Decides the next move depending on what is around and what has been currently explored, depending on the Nodes dictionary, choosing the smallest distance to the end 
    current_steps = nodes[pretty(a)]['s']
    next_step = current_steps + 1
    n_v,e_v,s_v,w_v = check_around(maze,a)
    if n_v == 0:
        find = pretty(find_north(a))
        if find not in nodes:
            nodes[pretty(find_north(a))] = {'co': find_north(a), 's': next_step, 'explored':False, 'via':pretty(a) ,'y-distance': distance(maze,find_north(a))[0],'x-distance': distance(maze,find_north(a))[1] }
        if find in nodes:
            if nodes[pretty(find_north(a))]['s'] > next_step:
                nodes[pretty(find_north(a))] = {'co': find_north(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_north(a))[0] , 'x-distance': distance(maze,find_north(a))[1]  }
    if e_v == 0:
        find = pretty(find_east(a))
        if find not in nodes:
            nodes[pretty(find_east(a))] = {'co': find_east(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_east(a))[0] , 'x-distance': distance(maze,find_east(a))[1]  }
        if find in nodes:
            if nodes[pretty(find_east(a))]['s'] > next_step:
                nodes[pretty(find_east(a))] = {'co': find_east(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_east(a))[0] , 'x-distance': distance(maze,find_east(a))[1] }
    if s_v == 0:
        find = pretty(find_south(a))
        if find not in nodes:
            nodes[pretty(find_south(a))] = {'co': find_south(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_south(a))[0] , 'x-distance': distance(maze,find_south(a))[1] }
        if find in nodes:
            if nodes[pretty(find_south(a))]['s'] > next_step:
                nodes[pretty(find_south(a))] = {'co': find_south(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_south(a))[0] , 'x-distance': distance(maze,find_south(a))[1] }
    if w_v == 0:
        find = pretty(find_west(a))
        if find not in nodes:
            nodes[pretty(find_west(a))] = {'co': find_west(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_west(a))[0] , 'x-distance': distance(maze,find_west(a))[1] }
        if find in nodes:
            if nodes[pretty(find_west(a))]['s'] > next_step:
                nodes[pretty(find_west(a))] = {'co': find_west(a), 's': next_step, 'explored':False,'via':pretty(a) ,'y-distance': distance(maze,find_west(a))[0] , 'x-distance': distance(maze,find_west(a))[1]  }
    nodes[pretty(a)]['explored'] = True
    return nodes

def find_lowest_steps(nodes):
    #Used to find the unexplored node with the lowest steps in the dictionary 
    lowest_steps = float('inf')
    for key in nodes:        
        y_d = nodes[key]['y-distance']
        x_d = nodes[key]['x-distance']
        total_d = y_d + x_d
        steps = nodes[key]['s'] + 20*((y_d/total_d)*y_d + (x_d/total_d)*x_d)
        done = nodes[key]['explored']
        if done == False:
            if steps < lowest_steps:
                lowest_steps = steps
                lowest_key = key
    return nodes[lowest_key]['co']

def check_end(maze,nodes):
    #Work out if the end node has been found in the dictionary
    finish = False
    end = pretty(find_end(maze))
    if end in nodes:
        finish = True
    return finish

def find_next(maze,nodes):
    return find_lowest_steps(nodes)
    
 
def find_path(maze,nodes):
    #After finding the end, go back and determine the fastest path depending on the nodes dictionary
    current = pretty(find_end(maze))
    path = []
    start = pretty(find_start(maze))
    finish = False
    while finish == False:
        path.append(current)
        if current == start:
            finish = True
            break
        current = nodes[current]['via']
    return path
        
def solve_maze(maze):
    #Incorporate all the functions in one central function to solve the maze, returning the best path to the end
    nodes = create_dict(maze)
    finish = False
    while finish == False:
        nodes = next_move(maze,nodes,find_lowest_steps(nodes))
        if check_end(maze,nodes) == True:
            finish = True
            break
    path = find_path(maze,nodes)
    return path,nodes

def show_path(maze):
    #Used as a visualisation of the maze to transfer to an excel, with the fastest path shown as "2"s
    path,nodes = solve_maze(maze)
    for i in range(len(path)):
        path_node = nodes[path[i]]['co']
        maze[path_node[0]][path_node[1]] = 2
    return maze
  
    
#Excel output part


def to_maze(input_file,read):
    #Analyse maze from the excel input
    read = read_file(input_file)
    read = read.replace(",","")
    use = read[:-1].split('\n')
    maze = []
    for i in range(len(use)):
        line = []
        for j in range(len(use[i])):
            line.append(int(use[i][j]))
        maze.append(line)
    return maze

def read_file(input_file):
    #Function to read excel file and return content
    filename = input_file
    file = open(filename,'r')
    read = file.read()
    file.close()
    return read


def write_file_output(output_file,path_excel):
    #Function to write the final path to an excel
    filename = output_file
    #output = 'maze_output.csv'
    file = open(filename,'w')
    file.write(path_excel)
    file.close()
    return

def to_excel(path):
    #Turns the path into a string to put into the excel
    path_excel = ''
    for i in range(len(path)):
        for j in range(len(path[i])):
            path_excel += str(path[i][j])
            if j != len(path[i]) - 1:
                path_excel += ','
            if j == len(path[i]) - 1:
                path_excel += '\n'
    return path_excel

def main():
    #Final main function that shoud be called to find path and show in the excels
    input_file = 'maze_input.csv'
    output_file = 'maze_output.csv'
    read = read_file(input_file)
    path = show_path(to_maze(input_file,read))
    write_file_output(output_file,to_excel(path))
    path2,nodes = solve_maze(to_maze(input_file,read))
    return
```
